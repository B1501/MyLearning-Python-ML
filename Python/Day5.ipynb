```python
x=np.linspace(0,5,11)

x
```

```python
print(x)
y=x**(10)
print(y)
plt.plot(y)
plt.show()
```

```python
plt.plot(x,y,'k*')
plt.show()
plt.plot(x,y,color = "r")
```

```python
plt.figure()
plt.plot(x,y)
plt.title("Square of x")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid()
plt.axis([0,5,0,25])
plt.show()
```

```python
x=np.linspace(0,10,21)
print(x)
y=x**2
plt.figure(figsize=[5,5],dpi=100,facecolor='#ff7f0e')
plt.plot(x,y,color='b',alpha=1,linestyle='-.',\
         linewidth=4,marker='d',markersize=10)
```

```python
x=np.linspace(0,10,21)
print(x)
y=x**2
plt.figure(figsize=[5,5],dpi=100,facecolor='#ff7f0e')
plt.plot(x,y,color='b',alpha=0.75,linestyle='-.',\
         linewidth=4,marker='o',markersize=7)
```

```python
x=np.linspace(0,10,21)
print(x)
y=x**2
plt.figure(figsize=[10,5])
plt.subplot(1,2,1)
plt.plot(x,y,'r--*')
plt.subplot(1,2,2)
plt.plot(y,x,'g-*')
plt.show
```

```python
x=np.linspace(0,10,21)
y=np.linspace(0,5,21)
a=x**2
b=y**2
plt.figure(figsize=[15,10],facecolor='#ff7f0e')
plt.subplot(2,2,1)
plt.plot(a,x,'r--*')
plt.title("a vs x")
plt.xlabel('a')
plt.ylabel('x')
plt.subplot(2,2,2)
plt.plot(x,a,'g--*')
plt.title("x vs a")
plt.xlabel('x')
plt.ylabel('a')
plt.subplot(2,2,3)
plt.plot(b,y,'c--*')
plt.title("b vs y")
plt.xlabel('b')
plt.ylabel('y')
plt.subplot(2,2,4)
plt.plot(y,b,'k--*')
plt.title("y vs b")
plt.xlabel('y')
plt.ylabel('b')
```

```python
y=[3,3,2,2,1,1]
x=[5,2,2,5,5,2]
a=[6,6,9,9,6,9,9]
b=[1,3,3,2,2,2,1]
c=[10,12,11,11,10,12]
d=[1,1,1,3,3,3]
plt.axis([0,15,0,4])
plt.plot(x,y)
plt.plot(a,b)
plt.plot(c,d)
plt.show()
```

```python
plt.figure()
x=np.linspace(0,5,11)
y=x**2
y1=x**3
plt.plot(x,y,'--',color='r',label="power of 2")
plt.plot(x,y1,'-*',color='g',label="power of 3")
plt.axis([0,5,0,20])
plt.legend(loc=5)
plt.show()
```

# Scatter Plot

```python
x=[11,42,32,54,64,13,31,71,16]
y=[23,35,43,24,52,65,31,19,46]
x1=[22,45,13,22,45,63,51,3,1]
y1=[5,8,16,13,45,77,12,7,2]

plt.scatter(x,y,color='r',label="red",marker='o',s=200,alpha=0.8)
plt.scatter(x1,y1,color='c',label=" blue",marker='d',s=200,alpha=0.75)
plt.legend()
plt.grid()
plt.show()
```

# Bar Graphs

```python

import matplotlib.style
import matplotlib as mpl
mpl.style.use('_classic_test')

x=[11,42,32,54,64]
y=[23,35,43,24,52]
x1=[22,45,13,22,56]
y1=[5,8,16,13,45]

plt.bar(x,y,color='r',label="red",width=3,align='edge',alpha=0.8)
plt.bar(x1,y1,color='c',label=" blue",width=3,align='edge',alpha=0.8)
plt.legend(loc=2)
plt.grid()
plt.show()
```

```python
x=[11,42,32,54,64]
y=[23,35,43,24,52]
x1=[22,45,13,22,56]
y1=[5,8,16,13,45]

plt.barh(x,y,color='r',label="red",height=3,align='center',alpha=0.8)
plt.barh(x1,y1,color='c',label=" blue",height=3,align='center',alpha=0.8)
plt.legend(loc=4)
plt.grid()
plt.show()
```

# Histogram

```python
x=100+np.random.randn(100)
plt.hist(x,50,color='g',cumulative=True)
plt.xlabel('people')
plt.ylabel("frequency")
plt.grid()
```

```python
x=100+np.random.randn(10000)
plt.hist(x,50,color='g',cumulative=False)
plt.xlabel('people')
plt.ylabel("frequency")
plt.grid()
```
