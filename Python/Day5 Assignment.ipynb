# Random Matrix

```python
import numpy as np
k=3
a = np.zeros((k,k),dtype = int)
print("enter {0} elements".format(k*k))
for i in range(k):
    for j in range(k):
        a[i][j]=input()
print(a)
d=np.random.randint(1,20,size=(k,k))
print("rand=\n",d)
# x=a+d
x=np.add(a,d)
print("rand+given=\n",x)
```

# Operations on matrices

```python
 import numpy as np
k=3
a = np.zeros((k,k),dtype = int)
print("enter {0} elements".format(k*k))
for i in range(k):
    for j in range(k):
        a[i][j]=input()

b = np.zeros((k,k),dtype = int)
print("enter {0} elements".format(k*k))
for i in range(k):
    for j in range(k):
        b[i][j]=input()
print("a=\n",a)
print("b=\n",b)
print("addition=\n",a+b)
print("subtract=\n",a-b)
print('multiply=\n',a*b)
print("division=\n",a/b)
```

# Plotting of Combined signal

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')
t=np.arange(1,10,0.1)
a=np.sin(t)
plt.figure(figsize=[17,10],facecolor="#ff7f0e",linewidth=10,edgecolor='w' )
plt.subplot(2,3,1)
plt.plot(t,a,'c--')
plt.title("sine")
plt.xlabel('time')
plt.grid()
plt.ylabel('amplitude')
b=np.cos(t)

plt.subplot(2,3,2)
plt.plot(t,b,'r--')
plt.grid()
plt.title("cos")
plt.xlabel('time')
plt.ylabel('amplitude')
c=np.tan(t)

plt.subplot(2,3,3)
plt.plot(t,c,'g--')
plt.grid()
plt.title("tan")
plt.xlabel('time')
plt.ylabel('amplitude')

plt.subplot(2,3,4)
plt.plot(t,a,'c-')
plt.plot(t,b)
plt.grid()
plt.title("sin+cos")
plt.xlabel('time')
plt.ylabel('amplitude')

plt.subplot(2,3,5)
plt.plot(t,a)
plt.plot(t,c,'g--')
plt.grid()
plt.title("tan + sin")
plt.xlabel('time')
plt.ylabel('amplitude')

plt.subplot(2,3,6)
plt.plot(t,b)
plt.plot(t,c,'b--')
plt.grid()
plt.title("cosine + tan")
plt.xlabel('time')
plt.ylabel('amplitude')
```

```python
t=np.arange(0,5,0.1)
a=np.sin(t)
b=np.sin(t+90)
c=np.sin(t+180)
# plt.figure(figsize=[5,15],facecolor="#ff7f0e",linewidth=10,edgecolor='w' )
plt.plot(t,a,'k',label="phase 1")
plt.plot(t,b,'r',label="phase 2")
plt.plot(t,c,'b',label="phase 3")
plt.legend(loc=5)
plt.title("THREE PHASE WAVEFORM")
plt.xlabel('Time')
plt.grid()
plt.ylabel('Amplitude')
```

# Plotting Exponential Curves

```python
a=np.arange(0,3,0.01)
x=a**2
y=a**3
z=a**4
k=a**5
l=a**6
plt.plot(a,'b--',label="power 1",linewidth=3)
plt.plot(a,x,'k--',label="power 2",linewidth=3)
plt.plot(a,y,'r--',label="power 3",linewidth=3)
plt.plot(a,z,'b--',label="power 4",linewidth=3)
plt.plot(a,k,'y--',label="power 5",linewidth=3)
plt.plot(a,l,'c--',label="power 6",linewidth=3)
plt.grid()
plt.axis([0,3,0,100])
plt.legend(loc=2)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("exponential curves")
```
