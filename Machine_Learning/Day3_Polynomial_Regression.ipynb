# Polynomial Regression

```python
%matplotlib inline
import matplotlib.pyplot as plt
#import matplotlib.style
#import matplotlib as mpl
mpl.style.use('classic')
x=[[6],[8],[10],[14],[18],[20],[22],[26]]
y=[[7],[9],[13],[17.5],[18],[16],[14],[16.5]]
plt.figure()
plt.title("Pizza  price plotted against diameter")
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(x,y,"go")
plt.axis([0,25,0,25])
plt.grid(alpha=1)
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=\
train_test_split(x,y,test_size=0.2,random_state=100)

plt.figure()
plt.title("Pizza  price plotted against diameter")
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(x_train,y_train,"go")
plt.plot(x_test,y_test,"bo")
plt.axis([0,25,0,25])
plt.grid(alpha=1)
```

```python
%precision 2
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
print(x_train)
poly = PolynomialFeatures(degree=5)
QC=poly.fit_transform(x)
QC1=poly.fit_transform(x_train)# Quadratic Coefficient
QC2=poly.fit_transform(x_test)
QC[0:5]
```

```python
pmodel = LinearRegression()
print(QC1)
print(y_train)
pmodel.fit(QC1,y_train)
```

```python
print("Actual Outputs",y_test)
print("predicted values on test\n",pmodel.predict(QC2))
```
